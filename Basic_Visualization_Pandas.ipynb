```
#First we need to get an average of HIV prevalence for each country across the timeframe. 

years = merged.loc[: , "1979":"2011"]
merged['hiv_average'] = years.mean(axis=1)

pd.options.display.float_format = '{:.2f}'.format

```

```
#Max and Min Values for Each Continent

countries_high = merged.loc[merged.groupby('Continent')['hiv_average'].idxmax()]
countries_low = merged.loc[merged.groupby('Continent')['hiv_average'].idxmin()]

#Grabbing a selected country that is not in highs or lows for later use as well
additional = merged.loc[(merged['Country_Name'] == "United States") | (merged['Country_Name'] == "New Zealand") | (merged['Country_Name'] == "Spain") | (merged['Country_Name'] == "Cameroon") | (merged['Country_Name'] == "Nepal")]
additional['binary'] = 'Neither'

#Calculating continental averages across time
averages = merged.groupby(['Continent']).mean()

countries_high['binary'] = 'high'
countries_low['binary'] = 'low'

highs_lows = pd.concat([countries_high, countries_low, additional, averages], axis=0)
highs_lows

```

4. Create two scatter plots to show the data (i.e. each country/region) in

plot is the HIV estimated prevalence, and the horizontal axis is the corresponding year average HIV estimated prevalence in each continent, which
you calculated above. Using different color to show data from different
continent (1 point). If you found any interesting result from the charts,
explain it.


```
ninety = merged[['Country_Name', 'Continent', '1990']]
avg1 = ninety.groupby('Continent').mean().iloc[0]['1990']
avg2 = ninety.groupby('Continent').mean().iloc[1]['1990']
avg3 = ninety.groupby('Continent').mean().iloc[2]['1990']
avg4 = ninety.groupby('Continent').mean().iloc[3]['1990']
avg5 = ninety.groupby('Continent').mean().iloc[4]['1990']


ninety.loc[ninety.Continent == 'Africa', '1990a'] = avg1 
ninety.loc[ninety.Continent == 'Africa', 'continent'] = 1
ninety.loc[ninety.Continent == 'Americas', '1990a'] = avg2 
ninety.loc[ninety.Continent == 'Americas', 'continent'] = 2
ninety.loc[ninety.Continent == 'Asia', '1990a'] = avg3
ninety.loc[ninety.Continent == 'Asia', 'continent'] = 3
ninety.loc[ninety.Continent == 'Europe', '1990a'] = avg4
ninety.loc[ninety.Continent == 'Europe', 'continent'] = 4
ninety.loc[ninety.Continent == 'Oceania', '1990a'] = avg5
ninety.loc[ninety.Continent == 'Oceania', 'continent'] = 5


plot = ninety.plot.scatter('1990a', '1990', c='continent', cmap='viridis')
plot.set_xlabel('Average HIV Prevalence Per Continent in 1990')
plot.set_ylabel('HIV Prevalence in 1990')
```

```
ten = merged[['Country_Name', 'Continent', '2010']]
avg1 = ten.groupby('Continent').mean().iloc[0]['2010']
avg2 = ten.groupby('Continent').mean().iloc[1]['2010']
avg3 = ten.groupby('Continent').mean().iloc[2]['2010']
avg4 = ten.groupby('Continent').mean().iloc[3]['2010']
avg5 = ten.groupby('Continent').mean().iloc[4]['2010']


ten.loc[ten.Continent == 'Africa', '2010a'] = avg1 
ten.loc[ten.Continent == 'Africa', 'continent'] = 1
ten.loc[ten.Continent == 'Americas', '2010a'] = avg2 
ten.loc[ten.Continent == 'Americas', 'continent'] = 2
ten.loc[ten.Continent == 'Asia', '2010a'] = avg3
ten.loc[ten.Continent == 'Asia', 'continent'] = 3
ten.loc[ten.Continent == 'Europe', '2010a'] = avg4
ten.loc[ten.Continent == 'Europe', 'continent'] = 4
ten.loc[ten.Continent == 'Oceania', '2010a'] = avg5
ten.loc[ten.Continent == 'Oceania', 'continent'] = 5


plot = ten.plot.scatter('2010a', '2010', c='continent', cmap='viridis')
plot.set_xlabel('Average HIV Prevalence Per Continent in 2010')
plot.set_ylabel('HIV Prevalence in 2010')
plot.legend()
```
