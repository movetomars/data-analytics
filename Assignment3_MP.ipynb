```
# INFO 5502
# Assignment 3 - Graphing Pi with Functions and Random Selection
# Maia Petee
#######################

from datascience import *
import numpy as np
import matplotlib
matplotlib.use('Agg', warn=False)
%matplotlib inline
import matplotlib.pyplot as plt

```

```
def pi_sim(repetitions):
  np.random.seed(0) # Seeding the random number generator
  xs = make_array() # Create an array for our x values
  ys = make_array() # Create an array for our y values
  inside = make_array() # Create an array to tell us whether points are inside circle
  
  center_x = 0.5 # Defining Circle
  center_y = 0.5
  radius = 0.5
  c = plt.Circle((center_x, center_y),radius, color='g', fill=False) 
  s = plt.Rectangle((0,0), 1, 1, color='r', fill=False)
  
  hits = 0
  
  for i in np.arange(repetitions):
      x = np.random.rand(1, 1)
      xs = np.append(xs, x)
      y = np.random.rand(1, 1)
      ys = np.append(ys, y)
      
      if ((x - center_x)**2 + (y - center_y)**2) < radius**2:
        inside = np.append(inside, 'green')
        hits += 1
      else:
        inside = np.append(inside, 'red')


    
  chart = plt.scatter(xs, ys, s=1.5, color=inside)

  fig = plt.gcf()
  ax = fig.gca()
  ax.set_aspect('equal', 'box')
  ax.add_artist(c)
  ax.add_artist(s)
  
  pi_calc = (hits/repetitions)*4
  print("Our guess was", str(pi_calc)+", whereas pi is", str(round(np.pi,6))+". That's a variance of", str(round(abs(np.pi-pi_calc), 4))+".")
```
